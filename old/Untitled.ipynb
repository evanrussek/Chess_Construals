```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
N_pieces = 20
p_include = .3
rv = binom(N_pieces,p_include)
x = np.arange(1,21)
fig,ax = plt.subplots(1, figsize = (3,3), dpi = 100)
ax.plot(x,rv.pmf(x))
ax.set_xlabel('# Pieces')
ax.set_ylabel('Probability')
ax.set_title('Independently including pieces (p = .3)')

rv = binom(N_pieces,p_include)
x = np.arange(1,21)
fig,ax = plt.subplots(1, figsize = (3,3), dpi = 100)
ax.plot(x,np.exp(-.1*x))
ax.set_xlabel('# Pieces')
ax.set_ylabel('Probability')
ax.set_title('Linear cost in number of pieces')
ax.set_ylabel('Unnormalized probability')
```
